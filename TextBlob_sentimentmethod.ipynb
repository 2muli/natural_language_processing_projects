```python
from textblob import TextBlob
```

Sentiment analysis helps us find the emotional tone of text whether it’s positive, negative or neutral. TextBlob.sentiment() method has two key components:

polarity: shows if the text is positive, negative or neutral.Has a score between -1 (negative) and 1 (positive) showing how positive or negative the text is.
subjectivity: shows whether the text is more opinion-based or factual.Has a score between 0 (factual) and 1 (opinion-based) showing how subjective or objective the text is.

```python
text="I love Laikipia University" #is more opinion based and positive
blob=TextBlob(text)
sentiment=blob.sentiment
print(sentiment)
```

```python
text1="Mount Kenya is the largest mountain in Kenya" #It is factual and neutral
blob1=TextBlob(text1)
sentiment1=blob1.sentiment
print(sentiment1)
```

```python
text1="I hate lazziness" #It is opinion based and negative
print(TextBlob(text1).sentiment)
```

```python
#Mixed statement
mixedtext="Engineering is a good course. Though it is difficult"
blob2=TextBlob(mixedtext)
sentiment2= blob2.sentiment
print(sentiment2)
```
