Create a Corpus from a given Dataset.
A corpus is a collection of documents (text data).Creating coepus involves the following:
1.Load your Dataset
2.Preprocess the Dataset
3.Create a Dictionary
4.Create Bag of Words Corpus

```python
import os

# Create a dummy sample_file.txt for demonstration purposes
# In a real scenario, you would ensure your dataset file is present or provide the correct path.
file_content = """
This is the first document.
This is the second document, and it's a bit longer.
Another document for testing purposes.
"""
with open("sample_file.txt", "w", encoding="utf8") as f:
    f.write(file_content)

doc = open("sample_file.txt", encoding="utf8")
print("File 'sample_file.txt' opened successfully.")
```

```python
#Preprocess the Dataset
from gensim.utils import simple_preprocess

tokenized=[]
for sentence in doc:
  tokenized.append(simple_preprocess(sentence,deacc=True))
print(tokenized)
```

```python
#.Create a Dictionary
from gensim.corpora import Dictionary
#"Corpora" is the plural form of "corpus"
dic= Dictionary(tokenized)
print(dic)
```

Once we have the dictionary we can create a Bag of Word corpus using the doc2bow( ) function. This function counts the number of occurrences of each distinct word, convert the word to its integer word id and then the result is returned as a sparse vector.

```python
#Create Bag of Words Corpus
Bow_corpus= [dic.doc2bow(doc, allow_update=True) for doc in tokenized]
print(Bow_corpus)
```
