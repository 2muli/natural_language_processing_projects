
Tokenization Using Spacy

Implementation of Tokenization Using SpaCy.
We have :
    1.Blank Model Tokenization
    2.Displaying the Pipeline Components
    3.Tokenization with Part-of-Speech Tagging and Lemmatization

```python
#Blank Model Tokenization
import spacy
nlp=spacy.blank('en')
s=nlp("I love exploring new programming language during my free time")
for token in s:
  print(token)

```

```python
#.Displaying the Pipeline Components
nlp=spacy.load('en_core_web_sm')
print(nlp.pipe_names)
s=nlp("I love exploring new programming language during my free time")
for token in s:
  print(token)
```

```python
#Tokenization with Part-of-Speech Tagging and Lemmatization
nlp= spacy.load('en_core_web_sm')
s=nlp("I love exploring new programming language during my free time")
for token in s:
  print(token ,'|',
              spacy.explain(token.pos_),  #The spacy.explain() function provides a description of the POS tag for each token
            '|', token.lemma_)
```
