```python
from nltk import word_tokenize,sent_tokenize

s="I like learning programming. And also watching movies"
print(f"Word tokenization: {word_tokenize(s)}")
print(f"Sentence tokenization: {sent_tokenize(s)}")
```

```python
from nltk.stem import PorterStemmer

# Creating the object of the stemmer
porter= PorterStemmer()

print(porter.stem("play"))
print(porter.stem("playing"))
print(porter.stem("running"))
print(porter.stem("eating"))
print(porter.stem("eating"))
print(porter.stem("plays"))
#NB// Stemmers might not always result in semantically meaningful base words eg
print(porter.stem("studies"))
print(porter.stem("communication"))
print(porter.stem("better"))
```

```python
from nltk.stem import WordNetLemmatizer

#Object
lemmatizer= WordNetLemmatizer()

print(lemmatizer.lemmatize("plays",'v'))
print(lemmatizer.lemmatize("playing",'v'))
print(lemmatizer.lemmatize("running",'v'))
print(lemmatizer.lemmatize("eating",'v'))

#Lemmatizers always result in semantically meaningful base words e.g
print(lemmatizer.lemmatize("studies",'v'))
print(lemmatizer.lemmatize("communication",'v'))
print(lemmatizer.lemmatize("better",'a'))
```

```python
#Part of speech tagging(pos_tag)

from nltk import word_tokenize
from nltk import pos_tag

s="I like learning programming. And also watching movies"
tokenized_words = word_tokenize(s)
print(pos_tag(tokenized_words))
```

```python
from nltk import word_tokenize,pos_tag,ne_chunk

s="Joseph studied at Laikipia University"
tokenised_words=word_tokenize(s)
tagged_words=pos_tag(tokenized_words)
## Apply Named Entity Recognition
print(ne_chunk(tagged_words))
```
